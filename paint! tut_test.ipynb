```python
import pygame # install w pip
pygame.init()

# default game settings
fps = 60
timer = pygame.time.Clock() # ctrls how fast game runs
# screen dimensions
WIDTH = 800
HEIGHT = 600

# used to avoid error in for-loops for mouse clicks
active_size = 0
active_color = 0
painting = []

# stating dimensions of game screen
screen = pygame.display.set_mode([WIDTH, HEIGHT]) 
pygame.display.set_caption('Paint!') # name of game

# while the game is running 
# active while loop



def draw_menu(size, colors):
    # banner displaying game menu (colors, brushes, etc)
    pygame.draw.rect(screen, 'gray', [0, 0, WIDTH, 70])
    
    pygame.draw.line(screen, 'black', (0, 70), (WIDTH, 70), 3)
    # args: what to draw on, what color, (x,y) start coords,
    # (x,y) end coords, width of line (px)
    
    # creating brush menu
       # draw rectangle = bg of brush select
       # white = radius of brush
    
    # extra large:
    xl_brush = pygame.draw.rect(screen, 'black',  
                                [10,10,50,50]) # position
    pygame.draw.circle(screen, 'white', (35,35), 20) # x,y, radius
    
    # large:
    l_brush = pygame.draw.rect(screen, 'black', 
                               [70,10,50,50])
    pygame.draw.circle(screen, 'white', (95,35), 15)
    
    # medium:
    m_brush = pygame.draw.rect(screen, 'black', 
                               [130,10,50,50])
    pygame.draw.circle(screen, 'white', (155,35), 10)
    
    # small
    s_brush = pygame.draw.rect(screen, 'black', 
                               [190, 10, 50, 50])
    pygame.draw.circle(screen, 'white', (215, 35), 5)
    
    # store brushes under brush_list
    brush_list = [xl_brush, l_brush, m_brush, s_brush]
    
    
    
    # lets you know what color you have selected
    
    pygame.draw.circle(screen, colors, (400, 35),30) 
    
    # outside border for color select
    pygame.draw.circle(screen, 'dark gray', (400, 35),30, 3)
    
    # selection borders for dif brush sizes
    if size == 20:
        pygame.draw.rect(screen, 'green', [10,10,50,3])
    elif size == 15:
        pygame.draw.rect(screen, 'green', [70,10,50,3])
    elif size == 10:
        pygame.draw.rect(screen, 'green', [130,10,50,3])
    elif size == 5:
        pygame.draw.rect(screen, 'green', [190,10,50,3])



    # need to define color vars for color menu to appear
    blue = pygame.draw.rect(screen, (0,0,255), # position
                            [WIDTH - 35, 10, 25, 25]) # (r,g,b) vals
                                                 # start(x,y), end(x,y)
            
    red = pygame.draw.rect(screen, (255,0,0),
                           [WIDTH - 35, 35, 25, 25]) # position horizontally
                                                # move 25 px for every 2 colors
    
    green = pygame.draw.rect(screen, (0,255,0),
                             [WIDTH - 60, 10, 25, 25])
    
    yellow = pygame.draw.rect(screen, (255,255,0),
                              [WIDTH - 60, 35, 25, 25])
    
    teal = pygame.draw.rect(screen, (0,255,255),
                            [WIDTH - 85, 10, 25, 25])
    
    purple = pygame.draw.rect(screen, (255,0,255),
                            [WIDTH - 85, 35, 25, 25])
    
    white = pygame.draw.rect(screen, (0,0,0),
                            [WIDTH - 110, 10, 25, 25])
    
    black = pygame.draw.rect(screen, (255,255,255),
                            [WIDTH - 110, 35, 25, 25])
    
    color_rect = [blue, red, green, yellow, teal, purple, 
                  white, black]
    rgb_list = [(0,0,255),(255,0,0),(0,255,0),(255,255,0),
               (0,255,255),(255,0,255),(0,0,0),(255,255,255)]
    
    # create brush menu
    return brush_list, color_rect, rgb_list


def draw_painting(paints):
    for i in range(len(paints)):
        pygame.draw.circle(screen, paints[i][0], paints[i][1], 
                          paints[i][2])

run = True
while run:
    
    
    # keeps program running, screen bg
    
    timer.tick(fps) # ticks = milliseconds
    # if fps = 60, then 60 miliseconds
    screen.fill('white') # use colors included in pygame  
    # or rgb vals
    
    
    # tracking mouse stuff
       # mouse = (x,y) coord, y = mouse[1]
    
    mouse = pygame.mouse.get_pos() # returns (x,y) tuple of position
    # of mouse
    left_click = pygame.mouse.get_pressed()[0] # tells us if 
    # left button clicked
    
    # show color selected w mouse cursor
    if left_click and mouse[1] > 70:
        painting.append((active_color, mouse, active_size)) # tells funct
        # info of circle needed to draw it
        # creates list every time mouse is clicked
    draw_painting(painting)
    
    # create mouse cursor, shows preview of mouse
    if mouse[1] > 70:
        pygame.draw.circle(screen, active_color, mouse, active_size)

    brushes, colors, rgb = draw_menu(active_size, active_color)
    
    # handles everything in your game
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False # stop running game
            
        if event.type == pygame.MOUSEBUTTONDOWN: # left/right click
            for i in range(len(brushes)): # for any # of brushes in list
                if brushes[i].collidepoint(event.pos): # if any point collides w mouse
                    # if any brush is being used on screen
                    # size of brush remains consistent w brush sizes
                    # in brush menu
                    active_size = 20 - (i * 5) # formula changes on brush size
                    # if loop starts w i=0 in brush list
                    # so for xl brush, i=0 
                    # -> outputs brush size = 20
            
            for i in range(len(colors)): # now for colors
                if colors[i].collidepoint(event.pos): 
                    active_color = rgb[i] 
                
    
    pygame.display.flip()
pygame.quit() # this doesnt work too well
```
